```python
import pandas as pd
import numpy as np

df=pd.read_csv('/content/SAHeart.csv')
```

```python
df.columns
```

```python
df.drop(["row.names","famhist"], axis=1, inplace=True)
```

```python
X=df.drop("chd",axis=1)
y=df.chd
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4444)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc= DecisionTreeClassifier(criterion="gini", max_depth=3,random_state=44444)
```

```python
dtc.fit(X_train,y_train)
```

```python
y_pred=dtc.predict(X_test)
```

```python
from sklearn.tree import export_text
```

```python
X_train.columns
```

```python
print(export_text(dtc,feature_names=[i for i in X_train.columns]))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_pred,y_test)

from sklearn.metrics import accuracy_score,precision_score,recall_score
print(accuracy_score(y_pred,y_test))
print(precision_score(y_pred,y_test))
print(recall_score(y_pred,y_test))

from sklearn.metrics import classification_report
print(classification_report(y_pred,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc= RandomForestClassifier(n_estimators=150,criterion="gini",max_depth=6)
```

```python
rfc.fit(X_train,y_train)
```

```python
y_pred= rfc.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_pred,y_test)

from sklearn.metrics import accuracy_score,precision_score,recall_score
print(accuracy_score(y_pred,y_test))
print(precision_score(y_pred,y_test))
print(recall_score(y_pred,y_test))

from sklearn.metrics import classification_report
print(classification_report(y_pred,y_test))
```
